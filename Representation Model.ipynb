```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from simpletransformers.language_representation import RepresentationModel
from sklearn.metrics.pairwise import cosine_similarity

import nltk
import string
import numpy as np
import pandas as pd
import re
import logging
```

```python
df = pd.read_csv('ML Team 6 Data Processing - sample bagofwords.csv')
df
```

```python
# 16 seconds for 400 topics = 1500 topics a minute

model = RepresentationModel("bert", "bert-base-uncased", use_cuda=False)

word_vectors = model.encode_sentences(df['Bag of Words'], combine_strategy='mean')
```

```python
word_vectors[0]
```

```python
def recommendations(title):

    cosine_similarities = cosine_similarity(word_vectors, word_vectors)
    
    topics = df['Topic']
    
    indices = pd.Series(df.index, index = df['Topic']).drop_duplicates()
    
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    sim_scores = sim_scores[0:6]
    
    for i in sim_scores:
        print(str(i) + '   ' + topics[i[0]])
```

```python
recommendations('Pandas add column from one dataframe to another based on a join')

# Joining arrays within group by clause
# Pandas add column from one dataframe to another based on a join

# only works for actual given titles
```
